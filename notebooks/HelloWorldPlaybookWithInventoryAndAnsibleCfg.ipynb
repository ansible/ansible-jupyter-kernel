# Hello World Playbook With Inventory and Ansible.cfg
* Defines an ansible.cfg 
* Defines an inventory
* Defines a play
* Defines a task

* Press shift-enter four times to run the play and task

```ansible
#ansible.cfg
[defaults]
host_key_checking=False
```

```ansible
#inventory
[all]
localhost
```

```ansible
#play
name: hello world
hosts: localhost
gather_facts: false
```

```ansible
#task
debug:
 msg: Hello World
```
